```python
import time
from bs4 import BeautifulSoup
```

```python
url = 'https://www.sfu.ca/computing/people/faculty.html'
response=requests.get(url)
```

```python
soup = BeautifulSoup(response.content)
```

```python
prof=soup.find_all('div',attrs={'class','textimage section'})

```

```python
import pandas as pd
```

```python
table = pd.DataFrame(columns = ['Name', 'Designation', 'Field_of_interest'])

for number, alloted in enumerate(prof):
    temp_val=alloted.find_all('h4')
    name=temp_val[0].text.split(',')[0]
    title=temp_val[0].text.split(',')[1]
    allotedarea=alloted.find_all('p')
    areas=allotedarea[0].text.split(':')
    interested_areas = None
    if(len(areas)==2):
        interested_areas=areas[1]
    table.loc[number] = [name, title, areas]
```
